```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('kc_house_data.csv')
df
```

```python
df.loc[:,'date'] = pd.to_datetime(df['date'])
df = df.set_index(['id','date'])
df
```

Roadmap:

Too many variables, with too many mixed types, making it hard to cluster

    Mixed types makes it harder to decide which distance metric to use, to define similarity between samples
    Too many variables makes the interpretation of the clusters harder, after grouping them: Which variables have more impact? How these clusters are distinct?
    
Therefore, a variable selection is in order:

    There is no universal way of selection the clusters' variables
    For any subset we choose, the method will always return a clustering option
        Does it makes sense for our question?

Here, we use a classification model to select the most important variables
    
    The goal is not to find the best prediction variable ever.
    We want to find a prediction variable which will help us decide the direction we want our clustering to head towards.
        This method is called supervisionized clustering.

A valid way, and recommended even, is to inquire the business on which direction makes more sense, and handpick the variables in this fashion.

Part 1: Classification Model

```python
X = df.drop(['price','sqft_basement','yr_renovated','zipcode','lat','long'],axis=1)
y = df.price
X
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_ts, y_tr, y_ts = train_test_split(X,y, random_state=61658)
X_tr.shape, X_ts.shape
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

params = {
    'max_leaf_nodes': np.unique( np.logspace(2,12,20,base=2,dtype=int)),
}

grid = GridSearchCV(
    DecisionTreeRegressor(criterion='mae', random_state=61658),
    params,
    scoring='neg_mean_absolute_error',
    cv=10,
    verbose=10,
    n_jobs=6,
)

grid.fit(X_tr, y_tr)

from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_ts, grid.predict(X_ts))
```
